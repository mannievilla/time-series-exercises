```python
i = 1
people_list = []


while i < 10:
    
    peeps_data = requests.get(f'https://swapi.dev/api/people/?page={i}')
    
    people_list.append(peeps_data)
    
    i += 1
    
```

```python
people_list[0].json()['results']
```

```python
people_list
```

```python
dict_list = []

for dicti in people_list:
    
    people_page = pd.DataFrame(dicti.json()['results'])
    dict_list.append(people_page)
    
```

```python
dict_list[1].head()
```

```python
df = pd.concat(dict_list, axis=0, ignore_index=True)
```

```python
df.head()
```

```python
len(df.homeworld.value_counts().to_list())
```
